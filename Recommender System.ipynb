```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
columns=["user_id","item_id","rating","timestamp"]
data = pd.read_csv('u.data',sep="\t",names=columns)
data.head()
```

```python
movie_titles=pd.read_csv('Movie_Id_Titles')
movie_titles.head()
```

```python
data=pd.merge(data,movie_titles,on='item_id')
data.head()
```

```python
data.groupby('title')['rating'].mean().sort_values(ascending=False).head(10)
```

```python
data.groupby('title')['rating'].count().sort_values(ascending=False).head(10)
```

```python
rating = pd.DataFrame(data.groupby('title')['rating'].mean())
rating['Total Reviews']= pd.DataFrame(data.groupby('title')['rating'].count())
rating.head()
```

```python
movie_mat = data.pivot_table(index='user_id',columns='title',values='rating')
movie_mat.head()
```

```python
user_rating = movie_mat['Star Wars (1977)']
similar_movie=movie_mat.corrwith(user_rating)
```

```python
corr_movie = pd.DataFrame(similar_movie,columns=['Correlation'])
corr_movie.dropna(inplace=True)
```

```python
corr_movie = corr_movie.join(rating['Total Reviews'])
corr_movie.head()
```

```python
corr_movie[corr_movie['Total Reviews']>100].sort_values('Correlation',ascending=False).head()
corr_movie.head()
```
